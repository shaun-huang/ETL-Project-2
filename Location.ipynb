```python
import pandas as pd

# Read csv file
df = pd.read_csv("Data/clean_data/clean_data.csv")
df.head()
```

```python
df1=df.copy()
df1['victim_age'] = df1['victim_age'].astype(int)
df1['victim_gender']=df1['victim_gender'].map(lambda x: x[0])
df1['victim_age']= df1['victim_age'].apply(str)
df1['Demographic_ID'] = df1[['victim_age', 'victim_gender','victim_race']].apply(lambda x: ''.join(x), axis=1)
df1.head()
```
